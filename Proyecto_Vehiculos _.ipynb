```python
#IMPORTANDO LIBRERIAS NECESARIAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generar un DataFrame con los datos del fichero
df=pd.read_csv('data.csv')
df
```

```python
#consultando columnas
print(df.columns)
```

```python
#renombrando campos
df = df.rename(columns={
    'Make': 'Marca',
    'Model': 'Modelo',
    'Year': 'Año',
    'Engine Fuel Type': 'Tipo de combustible',
    'Engine HP': 'Potencia HP',
    'Engine Cylinders': 'Cilindros del motor',
    'Transmission Type': 'Tipo de transmisión',
    'Driven_Wheels': 'Tracción',
    'Number of Doors': 'Número de puertas',
    'Market Category': 'Categoría de mercado',
    'Vehicle Size': 'Tamaño del vehículo',
    'Vehicle Style': 'Estilo del vehículo',
    'highway MPG': 'Rendimiento carretera (MPG)',
    'city mpg': 'Rendimiento ciudad (MPG)',
    'Popularity': 'Popularidad',
    'MSRP': 'Precio sugerido (MSRP)'
})
```

```python
#eliminando Valores Vacios
df.dropna()
```

```python
#Validando Datos Atipicos
plt.boxplot(df['Popularidad'],vert=False)
plt.show()
```

```python
#Primer Cuartil
Q1 = df ['Popularidad'].quantile(0.25)
print('Primer Cuartil', Q1)
```

```python
#Tercer Cuartil
Q3 = df['Popularidad'].quantile(0.75)
print('Tercer Cuartil', Q3)
```

```python
#Rango Intercuartil
IQR = Q3 - Q1
print('Rango Intercuartil', IQR)
```

```python
#calculo del bigote inferior
BI_Calculado=(Q1-1.5*IQR)
print ("BI_Calculado \n", BI_Calculado)
```

```python
#calculo del bigote superior
BS_Calculado=(Q3+1.5*IQR)
print ("BS_Calculado \n", BS_Calculado)
```

```python
#verificacion ouliers segun rangos calculados
ubicacion_outliers=(df['Popularidad']<BI_Calculado)|(df['Popularidad']>BS_Calculado)
print ("\n Ubicación de Outliers Vn", ubicacion_outliers)
```

```python
#valores sin outliers segun rangos entregados
ubicacion_sin_outliers=(df['Popularidad']>=BI_Calculado) & (df['Popularidad']<=BS_Calculado)
sin_outliers=df[ubicacion_sin_outliers]
print ("\n Ubicación sin  Outliers Vn", ubicacion_sin_outliers)
```

```python
#grafico de caja con valores sin outiers
plt.boxplot(sin_outliers['Popularidad'],vert=False)
plt.show()
```

```python
#diagrama por Mraca
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Marca', palette='ocean')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#relacion potencia -cilindro motor
sns.lmplot (x='Cilindros del motor', y='Potencia HP', data=df, fit_reg=True)
```

```python

# valores por marca
serie=df.Marca.value_counts()
plt.figure(figsize=(5, 10))
serie.plot.barh()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Marca', y='Rendimiento ciudad (MPG)', palette='crest', estimator='mean')
plt.xticks(rotation=90)
plt.ylabel('Promedio Rendimiento Ciudad (MPG)')
plt.title('Promedio de rendimiento en ciudad por marca')
plt.tight_layout()
plt.show()
```
